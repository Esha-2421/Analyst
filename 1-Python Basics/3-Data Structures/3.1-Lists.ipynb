```python
## Accessing list elements

fruits=["apple","banana","cherry","date","elderberry"]
print(fruits[0])  # First element
print(fruits[2])  # Third element
print(fruits[-1]) # Last element
```

```python
print(fruits[1:4])  # Slicing from index 1 to 3
print(fruits[1:])  # Slicing from index 1 to end
print(fruits[-1:-3])  # Slicing from last to third last (empty)
```

```python
## List comprehension

Basics Syntax [expression for item in iterable]

with conditional logic [expression for item in iterable if condition]

Nested List Comprehension [expression for item1 in iterable1 for item2 in iterable2]
```

```python
## Basic list comprehension

square=[num**2 for num in range(10)]
print(square)
```

```python
## list comprehension with condition

lst=[]
for i in range(10):
    if i%2==0:
        lst.append(i)
print(lst)
```

```python
even_numbers=[num for num in range(10) if num%2==0]
print(even_numbers)
```

```python
## Nested list comprehension

list1=[1,2,3,4]
list2=['a','b','c','d']
pair=[[i,j] for i in list1 for j in list2]
print(pair)
```

```python
## list comprehension with function calls

words=["hello","world","python","list","comprehension"]
lengths=[len(word) for word in words]
print(lengths)
```
