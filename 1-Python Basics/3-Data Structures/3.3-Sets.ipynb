```python
## create a set

my_set={1,2,3,4,5}
print(my_set)
print(type(my_set))
```

```python
my_set=set([1,2,3,4,5,6])
print(my_set)
```

```python
my_empty_set=set([1,2,3,6,5,4,5,6])
print(my_empty_set)
```

```python
## basics set operation
## Adding and removing elements

my_set.add(7)
print(my_set)
my_set.add(7)
print(my_set)
```

```python
## Remove the elements from a set

my_set.remove(3)
print(my_set)
```

```python
my_set.remove(10)
print(my_set)
```

```python
my_set.discard(11)
print(my_set)
```

```python
## pop method

removed_element=my_set.pop()
print(removed_element)
print(my_set)
```

```python
# clear all the elements

my_set.clear()
print(my_set)
```

```python
## Set membership test
my_set={1,2,3,4,5}
print(3 in my_set)
print(10 in my_set)
```

```python
## mathematical operation

set1={1,2,3,4,5,6}
set2={4,5,6,7,8,9}

## union

union_set=set1.union(set2)
print(union_set)

## Intersection

intersection_set=set1.intersection(set2)
print(intersection_set)

set1.intersection_update(set2)
print(set1)
```

```python
set1={1,2,3,4,5,6}
set2={4,5,6,7,8,9}

## difference

print(set1.difference(set2))
```

```python
## sets method

set1={1,2,3}
set2={3,4,5}

## is subset

print(set1.issubset(set2))
print(set1.issuperset(set2))
```

```python
## Counting unique words in text

text="In this tutorial we are discussing about sets"
words=text.split()

## convert list of words to set to get unique words

unique_words=set(words)
print(unique_words)
print(len(unique_words))
```
